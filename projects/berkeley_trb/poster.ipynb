```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels = ['Exposed\nvehicles', '\nETE (hr)', '\nExpose time\n(hr)', 'Distance from\nfire (miles)']#, 'M5', 'M6']
max_values = [2500, 4, 5100, 1.0]#, 0.1, 0.1]
baseline = [782, 2.9, 954, 0.6, 0.1, 0.1]
# f0_5 = [342, 2.8, 350, 0.8]
f2 = [1209, 2.8, 1569, 0.5, 0.1, 0.1]
dl = [729, 2.7, 910, 0.4, 0.1, 0.1]
dm = [747, 3.1, 906, 0.6, 0.1, 0.1]
# t0 = [757, 2.8, 885, 0.7, 0.1, 0.1]
# t0_25 = [821, 3.1, 906, 0.6, 0.1, 0.1]
vphh1 = [245, 1.8, 129, 0.9, 0.1, 0.1]
vphh3 = [2497, 4, 5057, 0.4, 0.1, 0.1]
# r0 = [943, 3.2, 1320, 0.6]
# r50 = [381, 2.1, 279, 0.8]
r100 = [86, 1.2, 35, 0.7, 0.1, 0.1]
p60 = [80, 4, 40, 0.7, 0.1, 0.1]
cl = [209, 1.7, 84, 0.9, 0.1, 0.1]
data = {'3 veh/hh': (vphh3, 'C3'), '1 veh/hh': (vphh1, 'C2'), 'Fast fire': (f2, 'C1'), 'Leave late': (dm, 'C0'),  'Rerouting': (r100, 'C4'), 'Phased': (p60, 'C5'), 'Contraflow': (cl, 'C7')}
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, polar=True)
angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
angles=np.concatenate((angles,[angles[0]]))
# for raw_stats in [s1, s2]:
for k, (v, color) in data.items():
    raw_stats = [v[i]/max_values[i] for i in range(len(max_values))]
    stats=np.concatenate((raw_stats,[raw_stats[0]]))
    ax.plot(angles, stats, 'o', c=color, ls='-', linewidth=2, label=k)

raw_stats = [baseline[i]/max_values[i] for i in range(len(max_values))]
stats=np.concatenate((raw_stats,[raw_stats[0]]))
ax.plot(angles, stats, 'o-', linewidth=3, c='k', label='Baseline')

ax.tick_params(axis='x', which='major', pad=30)
ax.set_thetagrids(angles[:-1] * 180/np.pi, labels)
# plt.tight_layout()
ax.set_yticklabels([])
ax.grid(True)
plt.legend(loc=[1.2,0.3])
```
